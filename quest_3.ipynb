```python
import numpy as np
import pandas as pd

wine = pd.read_csv('wineD.csv')
print(wine.describe())

wine = wine.apply(lambda x: (x - x.mean()) / x.std() , axis = 0)
print(wine.describe())

iris = pd.read_csv('iris.csv')
iris = iris.iloc[:,0 : 4]
print(iris.describe())

iris = iris.apply(lambda x: (x - x.mean()) / x.std() , axis = 0)
print(iris.describe())
```
